Question 2

```python
df2 = pd.get_dummies(df, 'Lead')
df2.head()
```

```python
df3 = df2.groupby(['Year'])['Lead_Male', 'Lead_Female'].sum()
df3['Male_Ratio'] = df3['Lead_Male']/(df3['Lead_Male']+df3['Lead_Female'])
print(df3.head())
plt.plot(df3['Male_Ratio'], 'ro')
plt.xlabel('Year')
plt.ylabel('Male Ratio')
plt.show()
```

Question 3

```python
df4= df[['Year','Lead','Number words female','Total words', 'Number of words lead', 'Gross']]
df4['Tot_female_words'] = np.where(df4['Lead']=='Female',df4['Number words female']+
                                   df4['Number of words lead'],df4['Number words female'])
df4['dominant'] = np.where((df4['Tot_female_words']/df4['Total words'])>0.5,'Female','Male')
df4[df4['dominant']=='Female']

df5 = df4.groupby('dominant')['Gross'].sum()
print(df5.to_string(),'\n\n')
```
